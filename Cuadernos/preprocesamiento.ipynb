# Preprocesamiento

```python
#
# Cargado de datos
import pandas as pd

direc = '../Datos/tasas_int.csv'
datos = pd.read_csv(direc, sep=',', decimal='.', index_col=0, parse_dates=True, )
datos.shape
```

```python
datos.head()
```

```python
#
# Renombrado con pyjanitor
from janitor import clean_names
df = datos.clean_names()
df['producto_de_credito'].unique()
```

```python
#
# Promedio de tasa_de_interes agrupado por nombre_entidad y by_month_fecha_corte

# Revisar la estructura de los datos
print("Columnas disponibles:")
print(df.columns)
```

```python
#
# Promedio de tasa de interés agrupado por nombre_entidad y by_month_fecha_corte

# Calcular promedio agrupado por los dos campos categóricos
promedio_agrupado = df.groupby(['nombre_entidad', 'by_month_fecha_corte',
                                'tipo_de_persona', 'tipo_de_garantia',
                                'producto_de_credito', 'tipo_de_credito',
                                'plazo_de_credito'])['tasa_efectiva_promedio_ponderada'].mean()

print("Promedio de tasa de interés por entidad y mes:")
print(promedio_agrupado.head(15))

# Como DataFrame para mejor visualización
df_promedio = promedio_agrupado.reset_index()
print("\nComo DataFrame:")
print(df_promedio.head(15))

# Mostrar estadísticas básicas
print(f"\nNúmero total de grupos: {len(promedio_agrupado)}")
print(f"Promedio general: {promedio_agrupado.mean():.4f}")
print(f"Desviación estándar: {promedio_agrupado.std():.4f}")

df_promedio.head()
```

```python
df_promedio.drop_constant_columns
```

```python
# Mostrar algunos ejemplos de entidades específicas
print("\nEjemplos por entidad (primeras 3 entidades):")
entidades_unicas = df['nombre_entidad'].unique()[:3]
for entidad in entidades_unicas:
    print(f"\n{entidad}:")
    datos_entidad = df_promedio[df_promedio['nombre_entidad'] == entidad]
    print(datos_entidad.head())
```

```python
# Guardar DataFrame como CSV
df_promedio.to_csv('../Datos/df_promedio.csv', index=False)
print("✅ DataFrame guardado como 'df_promedio.csv' en la carpeta Datos")
print(f"📊 Guardados {len(df_promedio)} registros")
```
